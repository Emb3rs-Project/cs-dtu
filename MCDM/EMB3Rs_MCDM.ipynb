```python
# Import the Pairwise class from pyanp
from pyanp.pairwise import Pairwise

# Create a list of alternatives
alternatives = [
    "Selling excess heat",
    "Converting excess heat",
    "Delivering excess heat",
    "Reintegrating excess heat",
    "Storing excess heat",
    "Using excess heat for cooling",
    "Using excess heat for drying",
    "Using excess heat for water heating",
    "Using excess heat for greenhouse heating",
]

# Create a Pairwise object with the alternatives as the index
pw = Pairwise(alternatives)

# Fill the pairwise comparison matrix with your judgments
# Use the set() method to assign a value to a pair of alternatives
# For example, pw.set("Selling excess heat", "Converting excess heat", 3) means that selling excess heat is moderately more important than converting excess heat
# You can also use the fill() method to fill the matrix with random values for testing purposes
#pw.fill()

pw.set("Selling excess heat", "Converting excess heat", 3)



# Calculate the priority vector of the alternatives
# Use the priority_from_pairwise() method to compute the priorities
# This will return a pandas Series with the alternatives as the index and the priorities as the values
priorities = pw.priority_from_pairwise()

# Print the priority vector
print(priorities)
```

```python
# Create a AHP hierarchy tree
ahp_tree = AHPTree()

# Add Goal node
ahp_tree.add_child("", "Goal")

# Add Criteria nodes
for criterion, criterion_info in weights.items():
    print(criterion)
    ahp_tree.add_child("Goal", criterion)
    # Add Sub-criteria nodes
    for sub_criterion, sub_criterion_weight in criterion_info["sub_weights"].items():
        ahp_tree.add_child(criterion, sub_criterion)
        # Add Alternative nodes
        for alternative in alternatives:
            ahp_tree.add_child(sub_criterion, alternative)  # Assume equal importance for alternatives

# Set Criteria priorities
for criterion, criterion_info in weights.items():
    ahp_tree.priority_from_children(criterion, [criterion_info["sub_weights"][sub_criterion] for sub_criterion in ahp_tree.get_children_names(criterion)])

# Calculate the priorities of the alternatives
priorities = ahp_tree.priority_table()

# Print the priorities
print(priorities)
```

## 2. Generate the AHP Tree

```python
def generate_plantuml(weights, alternatives):
    plantuml_code = "@startuml\n"
    plantuml_code += "[Goal] --> [Fluid Properties]\n"
    for criterion, criterion_info in weights.items():
        if criterion != "Fluid Properties":
            plantuml_code += "[Goal] --> [{}]\n".format(criterion)
        for sub_criterion in criterion_info["sub_weights"].keys():
            plantuml_code += "[{}] --> [{}]\n".format(criterion, sub_criterion)
            for alternative in alternatives:
                plantuml_code += "[{}] --> [{}]\n".format(sub_criterion, alternative)
    plantuml_code += "@enduml"
    return plantuml_code


#print(generate_plantuml(weights, alternatives))
```
